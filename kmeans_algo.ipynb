```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
np.random.seed(1)
num_samples=15
X = np.random.rand(num_samples, 2)*10
```

```python
k = 3
```

```python
kmeans = KMeans(n_clusters = k)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
centroids=kmeans.cluster_centers_
```

```python
cluster1 = X[labels == 0]
cluster2 = X[labels == 1]
cluster3 = X[labels == 2]
```

```python
plt.scatter(cluster1[:, 0], cluster1[:, 1], color = 'red', label = 'Cluster 1')
plt.scatter(cluster2[:, 0], cluster2[:, 1], color = 'blue', label = 'Cluster 2')
plt.scatter(cluster3[:, 0], cluster3[:, 1], color = 'green', label = 'Cluster 3')
plt.scatter(centroids[:, 0], centroids[:, 1], marker = '*', s = 50, color = 'black', label = 'Centroids')
plt.xlabel('Feature 1')
plt.xlabel('Feature 2')
plt.title('KMeans clustering')
plt.legend()
plt.show()
```
